```python
schema_cust = "customer_id integer, customer_first_name string customer_last_name customer_zip integer "

df = spark.read.format("csv").option("header",True).option("inferSchema",True).load("/data/datasets/Tests_data/customer.csv")
```

```python
df.show()
```

```python
from pyspark.sql.functions import col,when

df_with_null = df.withColumn("customer_zip", when(col("customer_id")==19,None).otherwise(col("customer_zip")))
```

```python
df_with_null.show()
```

```python
# Handle BAD records - PERMISSIVE (Default mode)
schema_cust = "customer_id integer, customer_first_name string customer_last_name customer_zip integer bad_record string "

df_with_null = spark.read\
            .format("csv")\
            .option("header", True)\
            .option("mode", "PERMISSIVE")\
            .option("columnNameOfCorruptRecord", "bad_record")\
            .load("/data/datasets/Tests_data/customer.csv")
```

```python
df_with_null.printSchema()
```
