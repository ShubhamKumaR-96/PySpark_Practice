```python
# Get unique data
# Select distinct emp * from emp

emp_unique = emp.distinct()
```

```python
emp_unique.show()
```

```python
# Unique of departments_ids
# Select distinct dispartments_id from emp

emp_distinct_dep = emp_unique.select("department_id").distinct()
```

```python
emp_distinct_dep.show()
```

```python
# Window Functions
# Select *, from max(salary) over(partitions by department_id order by salary desc) as max_salary from emp_unique
from pyspark.sql.window import Window
from pyspark.sql.functions import max,col,desc

window_fun = Window.partitionBy(col("department_id")).orderBy(col("salary").desc())
max_func = max(col("salary")).over(window_fun)

emp_1 = emp.withColumn("max_salary",max_func)
```

```python
emp_1.show()
```

```python
# Window Functions - 2nd highest salary of each department
# Select *, from row_number over(partitions by department_id order by salary desc) as rn from emp_unique where rn = 2
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number,col,desc

window_fun = Window.partitionBy(col("department_id")).orderBy(col("salary").desc())

rn =  row_number().over(window_fun)

emp_2 = emp.withColumn("rn",rn).where("rn = 2")
```

```python
emp_2.show()
```

```python
# Window function using select expr
# Select *, row_number() over(partitionBy department_id order by salary desc) as rn from emp_unique where rn = 2
from pyspark.sql.functions import expr

emp_3 =  emp.withColumn("rn", expr("row_number() over(partition by department_id order by salary desc)")).where("rn = 2")
```
