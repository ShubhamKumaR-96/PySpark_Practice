```python
# Create emp dataframe

emp_df = spark.createDataFrame(emp_data,emp_schema)
```

```python
# Show data (Action)
emp_df.show()
```

```python
emp_df.printSchema()
```

```python
# Casting Column
# Select employee_id, name, age, cast(salary as double) as salary from emp

from pyspark.sql.functions import col,cast

emp_casted = emp_df.select("employee_id","name","age",col("salary").cast("double"))
```

```python
# Adding Columns
# Select employee_id, name, age, salary, (salary * 0.2) as tax from emp_casted

emp_taxed =  emp_casted.withColumn("tax",col("salary") * 0.2)
```

```python
emp_casted.printSchema()
```

```python
emp_taxed.show()
```

```python
# Literals
# select employee_id, name,age,salary,tax,1 as columnOne, 'two' as columns from emp_taxed
from pyspark.sql.functions import lit

emp_new_cols = emp_taxed.withColumn("columnOne",lit(1)).withColumn("Columntwo",lit("two"))
```

```python
# Show dataframe (ACTIONS)
emp_new_cols.show()
```

```python
# Renaming Columns
# Select employee_id as emp_id, name, age, salary,tax,ColumnOne,Columntwo from emp_new_cols

emp_renamed_cols = emp_new_cols.withColumnRenamed("employee_id","emp_id")

```

```python
# Show dataframe (Action)
emp_renamed_cols.count()
```

```python
# Column name with spaces
# Select employee_id as emp_id, name, age, salary, tax, columneOne, columntwo as Column Two from emp_renamed_cols

emp_rename_with_space = emp_renamed_cols.withColumnRenamed("Columntwo", "column Two")
```

```python
# Remove columns

emp_filterd=emp_rename_with_space.drop("columnOne","column two")
```

```python
emp_filterd.show()
```

```python
# Limit data
# Select emp_id, name, age, salary from emp_dropped

emp_limit = emp_filterd.limit(5)
```
