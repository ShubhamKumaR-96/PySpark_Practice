```python
emp = spark.createDataFrame(emp_data,emp_schema)
```

```python
emp.show()
```

```python
emp.printSchema()
```

```python
# Case when
# Select employee_id, name, age, salary, gender
# case when gender = "Male" the 'M', when gender="Female" the 'F' else null end as new_gender, hire_date from emp

from pyspark.sql.functions import when, col, expr

emp_fixed_gender = emp.withColumn("new_gender", when(col("gender") == "Male" , "M")\
                       .when(col("gender") == "Female", "F").otherwise(None))

emp_fixed_gender_1 = emp.withColumn("new_gender", expr("case when gender = 'Male' then 'M' when gender = 'Female' then 'F' else null end" ))
```

```python
emp_fixed_gender_1.show()
```

```python
# Replace in strings
# Select emplotyee_id, name, replace(name,'s','k') as new_name,age,salary, gender , new_gender, hire_date from emp_gender_fixed

from pyspark.sql.functions import regexp_replace

emp_name_fixed  =  emp_fixed_gender_1.withColumn("new_name",regexp_replace(col("name"),"S","k"))
```

```python
emp_name_fixed.show()
```

```python
# Convert Date
# Select *, to_date(hire_date,'YYYY-MM-DD') as hire_date from emp_name_fixed

from pyspark.sql.functions import to_date

emp_date_fix = emp_name_fixed.withColumn("hire_date",to_date(col("hire_date"),'yyyy-MM-dd'))
```

```python
emp_date_fix.printSchema()
```

```python
# Add DATE COLUMN
# ADD CURRENT_DATE, CURRENT_TIMESTAMP, EXTRACT YEAR FROM HIRE_DATE

from pyspark.sql.functions import to_date,current_date,current_timestamp

emp_dated = emp_date_fix.withColumn("date_now",current_date())\
                           .withColumn("ingestion_date",current_timestamp())
```

```python
emp_dated.show(truncate=False)
```

```python
# Drop nall value
emp_1 = emp_dated.na.drop()
```

```python
# Fix null values
# select *, nvl('new_gender','0') as new_gender from emp_dated

from pyspark.sql.functions import coalesce, lit

emp_null_df = emp_dated.withColumn("new_gender",coalesce(col("new_gender"),lit("0")))
```

```python
# Drop old columns and fix new columns
emp_final_df = emp_null_df.drop("name","gender").withColumnRenamed("new_name","name").withColumnRenamed("new_gender","gender")
```

```python
emp_final_df.show()
```

```python
emp_final_df.write.format("csv").save("/data/outputs/4/emp.csv")
```
