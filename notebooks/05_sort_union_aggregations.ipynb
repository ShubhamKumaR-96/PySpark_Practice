```python
# Create emp DataFrame

emp_data_1 = spark.createDataFrame(data=emp_data_1, schema=emp_schema)
emp_data_2 = spark.createDataFrame(data=emp_data_2, schema=emp_schema)
```

```python
# Show data 
emp_data_1.show()
emp_data_2.show()
```

```python
# Print Schema
emp_data_1.printSchema()
emp_data_2.printSchema()
```

```python
# UNION and UNION ALL
# Select * from emp_data_1 union select * from emp_data_2
emp =  emp_data_1.unionAll(emp_data_2)
```

```python
emp.show()
```

```python
# Sort the emp data based on desc salary
from pyspark.sql.functions import col, desc, asc

emp_sorted = emp.orderBy(col("salary").desc())
```

```python
emp_sorted.show()
```

```python
# Aggregation
from pyspark.sql.functions import count,sum
total_dept_df = emp_sorted.groupBy("department_id").agg(count("employee_id").alias("total_dept_count"))
                                                  
```

```python
total_dept_df.show()
```

```python
# Aggregations
# Select dept_id,sum(salary) as avg_dept_salary from emp_sorted group by dept_id having avg(salary) > 50000
from pyspark.sql.functions import count,sum
avg_salary_by_dept = emp_sorted.groupBy("department_id").agg(sum("salary").alias("total_dept_salary")).orderBy(col("total_dept_salary").desc())
```

```python
avg_salary_by_dept.show()
```

```python
# Aggregation with having clause
# Select dept_id, avg(salary) as avg_dept_salary from emp_sorted group by dept_id having avg(salary)>50000
from pyspark.sql.functions import avg
# avg_salary = emp_sorted.groupBy("department_id").agg(avg("salary").alias("avg_dept_salary")).filter(avg("salary") > 50000)
avg_salary_1 = emp_sorted.groupBy("department_id").agg(avg("salary").alias("avg_dept_salary")).where("avg_dept_salary > 50000")
```

```python
avg_salary_1.show()
```

```python
# UNION matches by POSITION, not column name!
df1 = spark.createDataFrame([(1, "a")], ["id", "name"])
df2 = spark.createDataFrame([(2, "b")], ["emp_id", "emp_name"])
```

```python
# This works but uses df1's column names — be careful!
df1.union(df2).show()
```

```python
# Safe approach — align by name first
df2_renamed = df2.toDF("id", "name")
df1.union(df2_renamed).show()
```

```python
df = df2.unionAll(df1)
```
